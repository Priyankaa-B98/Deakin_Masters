## STEP 3:

**Purpose of Step-3:**

To calculate the bmi of both male and female and add this as eighth column in both the datasets respectively.

The **BMIs(bmis)** is calculated by dividing the person's weight in kilograms by the square of their heights in meters.

```python
# Calculating the bmi of male:

male_weight = male[:, 0] # this column has the weight of the men's in kilograms.

male_height = male[:, 1] # this column has the height of the men's in centimeters (cm).

# To calculate the bmi of male:
male_bmi = (male_weight) / ((male_height / 100) ** 2)

# Here male_height that contains the value in centimeters (cm), is now divided by 100 and is converted into meters.
```

```python
# After calculating the BMI, this can be added as eighth column to the original dataset.

male = np.column_stack((male, male_bmi))

print("The shape of the dataset after adding the BMI as another column:", male.shape)
```

```python
# Calculating the bmi of female:

female_weight = female[:, 0] # this column has the weight of the female's in kilograms.

female_height = female[:, 1] # this column has the height of the female's in centimeters (cm).

# To calculate the bmi of male:
female_bmi = (female_weight) / ((female_height / 100) ** 2)

# Here male_height that contains the value in centimeters (cm), is now divided by 100 and is converted into meters.
```

```python
# After calculating the BMI, this can be added as eighth column to the original dataset.

female = np.column_stack((female, female_bmi))

print("The shape of the dataset after adding the BMI as another column:", female.shape)
```

### Conclusion of Step-3:

1. The columns of height of the female and male is set to female_height and male_height respectively.

2. The columns of weight of the female and male is set to female_weight and male_weight respectively.

3. In order to calculate the BMI, the heights are converted from centimeters to meters unit and is calculated by dividing the person's weight in kilograms by the square of their heights in meters.

4. These calculated bmi's for male and female are stored as eighth column to the original dataset.

**Purpose of Step-8:**

To draw a pairplot for the male heights, weights, waist circumferences, hip circumferences, and BMIs 5 columns only.

```python
def pairplot(X, labels, bins = 21, alpha = 0.1):
    assert X.shape[1] == len(labels)
    
    k = X.shape[1]
    fig, axes = plt.subplots(nrows=k, ncols=k, sharex = "col", sharey = "row",
        figsize = (plt.rcParams["figure.figsize"][0], )*2)
    for i in range(k):
        for j in range(k):
            ax = axes[i, j]
            if i == j:
                ax.text(0.5, 0.5, labels[i], transform = ax.transAxes, ha = "center", va = "center", size = "x-small")
            else:
                ax.plot(X[:, j], X[:, i], ".", color = "black", alpha = alpha)
  
```

```python
columns = ["height", "weight", "waist", "hip", "BMI"]
n = [1, 0, 6, 5, 7]
print(n)
print(columns)
pairplot(male[:, [1,0,6,5,7]], columns)
plt.show()
```

## STEP - 9:

To compute the Pearson's and Spearman's correlation coefficients for all the variables mentioned in subtask 8.

Also, to present these coefficients on 2 correlation heatmaps.

```python
def corrheatmap(R, labels, title = "Correlation Heatmap"):
    assert R.shape[0] == R.shape[1] and R.shape[0] == len(labels)
    k = R.shape[0]
    plt.imshow(R, cmap=plt.colormaps.get_cmap("RdBu"), vmin = -1, vmax = 1)

    # adding text labels:
    
    for i in range(k):
        for j in range(k):
            plt.text(i, j, f"{R[i, j]:.2f}", ha = "center", va = "center",
                color="black" if np.abs(R[i, j])<0.5 else "white")

    plt.xticks(np.arange(k), labels = labels, rotation = 30)
    plt.tick_params(axis = "x", which = "both",
        labelbottom = True, labeltop = False, bottom = False, top = False)

    plt.yticks(np.arange(k), labels = labels)
    plt.tick_params(axis = "y", which="both",
        labelleft = True, labelright = False, left = False, right = False)

    plt.grid(False)
    
# Adding the title
    plt.title(title)
```

```python
R = np.corrcoef(male, rowvar = False) # This is used to print the Pearson's coefficient.

P,_ = spearmanr(male, axis = 0) # This is used to print the Spearman's coefficient.

order = [1,0,6,5,7]

plt.subplot(1,2,1)
corrheatmap(R[np.ix_(order, order)], ['Height', 'Weight', 'Waist', 'Hip', 'BMI'], "Pearson's r")
plt.subplots_adjust(wspace = 1)

plt.subplot(1,2,2)
corrheatmap(P[np.ix_(order, order)], ['Height', 'Weight', 'Waist', 'Hip', 'BMI'], "Spearman's r")
plt.show()
```

### Conclusion of Step - 9:

1. A corrheatmap() function is created in order to print the heatmap of variables mentioned that is, Height, Weight, waist, Hip, BMI. Also based on two types of correlation coefficients, the heatmaps are plotted.

2. Pearson's coefficient assumes that variables are normally distributed or having linear relationships Whereas Spearman's coefficient assumes that variables have monotic relationships.

3. Weight is positively coorelated to BMI. Also waist and hip are postively correlated.

4. These correlation coefficient values describes how strong the association is between 2 variables.
